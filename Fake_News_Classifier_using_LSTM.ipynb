```
### Importing Tensorflow
import tensorflow as tf
tf.__version__
```

```
#### Importing Keras Libraries
from tensorflow.keras.layers import Embedding
### Use for add pading(before or after) to Lstm Embedding Layer
from tensorflow.keras.preprocessing.sequence import pad_sequences
####for converting sentences to one_hot representation given Vocabulary Size
from tensorflow.keras.preprocessing.text import one_hot
### for model
from tensorflow.keras.models import Sequential
#### Lstm and Dense Layer
from tensorflow.keras.layers import LSTM,Dense
```

```
### Vocabulary Size
vocab_size=5000
```

```
### making copy of x data
msg=x.copy()
```

```
msg.head()
```

```
### Reseting data because first we had deleted nan values
msg.reset_index(inplace=True)
```

```
msg.tail()
```

```
msg.head()
```

```
msg.shape
```

```
#### Importing nltk libraries
import nltk
import re
from nltk.corpus import stopwords ### used for removing unnecessary data from text.
```

```
nltk.download('stopwords')
```

```
#### Data Preprocessing

###Stemming:- is the process of reducing a word to its word stem that affixes to suffixes and prefixes or to the roots of words known as a lemma

### Stemming we used PorterStemmer
from nltk.stem.porter import PorterStemmer
porter=PorterStemmer()
corpus=[]
for i in range(0,len(msg)):
  print(i)
  review=re.sub('[^a-zA-Z]'," ",msg['title'][i]) #### replacing all values except a-zA-Z characters with " "(blank space)
  review=review.lower()  #### Putting all reviews in lower case
  review=review.split() ### splitting reviews in words for applying stopwords 

  review=[porter.stem(words) for words in review if not words in stopwords.words('english')] ### stemming done onwords which are not match with stopwords
  review=' '.join(review) #### joining of words are done here
  corpus.append(review) ### Appending 

```

```
corpus
```

```
len(corpus)
```

```
#### One_hot process
onehot_repr=[one_hot(word,vocab_size) for word in corpus] ###Giving each word from corpus list for one_hot 
onehot_repr
```

# **Embedding Representation**

```
sent_length=20
### giving length 20 for array rows
### padding added for making sentences equal length
#### padding='pre' means add padding prefix hence 0 added for remaining columns
embedded_docs=pad_sequences(onehot_repr,maxlen=sent_length,padding='pre')
embedded_docs
```

```
len(embedded_docs)
```

```
print(onehot_repr[0])

print(embedded_docs[0] )
```

```
#### Creating Model

### Embedding Layer take input and convert into specific number of Feature of vectors
embedding_vector_feature=40
model=Sequential()
### Embedding Layer
model.add(Embedding(vocab_size,embedding_vector_feature,input_length=sent_length))
### Adding Lstm layer
model.add(LSTM(100))
### Addin Dense Layer because its classification problem
model.add(Dense(units=1,activation='sigmoid')) #### units =1 because it is classification problem and return 1 value either 0 or 1
### Model Compilation 
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy']) 
### model Summary
model.summary()

```

```
#### Putting all embedded_docs values in x_final
x_final=np.array(embedded_docs)
y_final=np.array(y)
```

# **Model Training**

```
### Splitting Data

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x_final,y_final,test_size=0.33)
```

```
model.fit(x_train,y_train,validation_data=(x_test,y_test),epochs=10,batch_size=64)
```

```
#### Prediction

###performance metrics and accuracy 

### predict_classes use because our y_test data is in array format
y_pred=model.predict_classes(x_test)
```

```
y_pred
```

```
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)
```

```
cm
```

```
from sklearn.metrics import accuracy_score
acc=accuracy_score(y_test,y_pred)
acc
```

# **Adding Dropout Layer**

```
#### Creating Model
from tensorflow.keras.layers import Dropout

### Embedding Layer take input and convert into specific number of Feature of vectors
embedding_vector_feature=40
new_model=Sequential()
### Embedding Layer
new_model.add(Embedding(vocab_size,embedding_vector_feature,input_length=sent_length))
#### Adding Dropout
new_model.add(Dropout(0.3))
### Adding Lstm layer
new_model.add(LSTM(100))
#### Adding Dropout
new_model.add(Dropout(0.3))
### Addin Dense Layer because its classification problem
new_model.add(Dense(units=1,activation='sigmoid')) #### units =1 because it is classification problem and return 1 value either 0 or 1
### Model Compilation 
new_model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy']) 
### model Summary
new_model.summary()
```

```
new_model.fit(x_train,y_train,validation_data=(x_test,y_test),epochs=10,batch_size=64)
```

```
#### Prediction

###performance metrics and accuracy 

### predict_classes use because our y_test data is in array format
y_pred_new=new_model.predict_classes(x_test)
```
